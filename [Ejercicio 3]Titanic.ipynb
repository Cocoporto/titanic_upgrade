#### Primero una breve descripción de la base de datos: 
Las columnas del `dataset` son:

* `passengerId`: Tipo `int` - Valor de identificación único de cada pasajero.
* `name`: Tipo `string` - Nombre del pasajero.
* `sex`: Tipo `factor` - Género del pasajero (masculino o femenino).

* `sibsp`: Tipo `factor ordenado` - Número de hermanos/cónyuges a bordo del pasajero.
* `cabin`: Tipo `factor` - Tipo de cabina ocupada por el pasajero.
* `parch`: Tipo `factor ordenado` - Número de padres/hijos a bordo del pasajero.
* `survived`: Tipo `factor 2 de dos niveles` - Indica si el pasajero sobrevivió al hundimiento del Titanic (sí o no).

##### División de la base en grupos etarios

```python
#Ahora podriamos dividir a los pasajeros en 3 categorias: 
# niños= < 18 años
# adultos = 18 a 65 años
# adulto_mayor = > 65 años 
# y para los nulos marcarlos como Desconocido
def categorizar_edades(edades):
    categorias = []
    for edad in edades:
        if pd.isnull(edad):  
            categorias.append('Desconocido')  
        elif edad < 18:
            categorias.append('Niños')
        elif edad < 60:
            categorias.append('Adultos')
        else:
            categorias.append('Adulto Mayor')
    return categorias
df['Edad_Categoria'] = categorizar_edades(df['Age'])
print(df[['Age', 'Edad_Categoria']])
```

```python
# Calcula el rango intercuartílico para determinar los límites del filtro
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1 #IQR es el rango intercuartilico, es decir el 50% de los datos

# Calcula los límites inferior y superior para el filtro
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filtra los datos para excluir los outliers
filtered_data = df[(df['Fare'] >= lower_limit) & (df['Fare'] <= upper_limit)]

# Grafica la relación entre Pclass y Fare con los datos filtrados
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=filtered_data, palette='flare', hue='Pclass')
plt.title('Relación entre Pclass y Fare (Sin Outliers)')
plt.xlabel('Pclass (Clase)')
plt.ylabel('Fare (Tarifa)')
plt.show()
```
